### Face Recognition 

```python
import cv2
import numpy as np    #frontalface_default
```

```python
c = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
cap = cv2.VideoCapture(0)
while(cap.read()):
    res , frame = cap.read()
    #grey = cv2.cvtColor(frame , 0)
    det = c.detectMultiScale(frame)
    if(len(det)>0):
        print(det)
        (x,y,h,w) = det[0]
        frame = cv2.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),2)
        font = cv2.FONT_HERSHEY_SIMPLEX
        text = 'kamal'
        cv2.putText(frame,text,(x,y),font,1,(255,255,0),2)
            
    cv2.imshow('a',frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break 
    
    
cap.release()
cv2.destroyAllWindows()
```

### Smile Detection 

```python
c = cv2.CascadeClassifier('haarcascade_smile.xml')
```

```python
cap = cv2.VideoCapture(0)
while(cap.read()):
    res , frame = cap.read()
    #grey = cv2.cvtColor(frame , 0)
    det = c.detectMultiScale(frame,scaleFactor=1.3,minNeighbors=5)
    if(len(det)>0):
        print(det)
        (x,y,h,w) = det[0]
        frame = cv2.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),2)
            
    cv2.imshow('a',frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break 
    
    
cap.release()
cv2.destroyAllWindows()
```

### Real video 

```python
c = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
cap = cv2.VideoCapture('obama.mp4')
width= int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height= int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
forcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter('output3.mp4',forcc,8.0,(width,height))
while(cap.read()):
    res , frame = cap.read()
    #grey = cv2.cvtColor(frame , 0)
    det = c.detectMultiScale(frame)
    if(len(det)>0):
        print(det)
        (x,y,h,w) = det[0]
        frame = cv2.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),2)
        font = cv2.FONT_HERSHEY_SIMPLEX
        text = 'Barack Obama'
        cv2.putText(frame,text,(x,y),font,1,(255,255,0),2)
        out.write(frame)    
    cv2.imshow('a',frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break 
    
   
cap.release()
out.release()
cv2.destroyAllWindows()
```
